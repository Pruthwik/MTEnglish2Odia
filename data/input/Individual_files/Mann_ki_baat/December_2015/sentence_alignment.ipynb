```python
import re
```

```python
with open('en.txt', 'r+', encoding='utf-8') as enfile, open('or.txt', 'r+', encoding='utf-8') as orfile:
    en_file = enfile.read().split('\n')
    or_file = orfile.read().split('\n')
```

```python
for en_cnt, en_sent in enumerate(en_file):
    for or_cnt, or_sent in enumerate(or_file):
        if re.findall('\d+', en_sent):
            if re.findall('\d+', en_sent) == re.findall('\d+', or_sent):
                # if matched then pair up the first instance of en and or sentences and traverse to the next en_sentence
                print(en_cnt, or_cnt, en_sent, '||', or_sent)
                prev_or_cnt = or_cnt
                break
        elif re.findall('\w+', or_sent):
            if re.findall('\w+', en_sent) == re.findall('\w+', or_sent):
                # if matched then pair up the first instance of en and or sentences and traverse to the next en_sentence
                print(en_cnt, or_cnt, en_sent, '||', or_sent)
                print(re.findall('\w+', en_sent))
                prev_or_cnt = or_cnt
                break
#             # if not matched then print both the sentences as it is and move to next en_sentence
#             print(or_cnt, en_sent, '||', or_sent)
#             break
```
